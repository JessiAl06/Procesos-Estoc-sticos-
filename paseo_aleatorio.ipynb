```python
# Importamos librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Número de pasos
n_steps = 1000

# Generamos pasos aleatorios (+1 o -1)
steps = np.random.choice([-1, 1], size=n_steps)

# Calculamos la posición acumulada
position = np.cumsum(steps)

# Graficamos el paseo aleatorio
plt.figure(figsize=(10,6))
plt.plot(position, label="Paseo Aleatorio")
plt.axhline(0, color="red", linestyle="--", label="Referencia")
plt.title("Simulación de un Paseo Aleatorio")
plt.xlabel("Número de pasos")
plt.ylabel("Posición")
plt.legend()
plt.show()
```
